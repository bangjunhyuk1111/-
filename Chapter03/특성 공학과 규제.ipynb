# 사이킷런의 변환기

`변환기`란? 사이킷런은 특성을 새로 만들거나 전처리하기 위해 여러 클래스를 제공한다. 이것을 `변환기`라고 한다.

**fit()**, **transform()** 함수를 제공한다.

# LinearRegression(회귀 훈련)과 PolynomialFeatures(특성 조작) 에서 fit() 메소드

함수명은 같지만 이 둘의 쓰임은 완전히 다르다.

`LinearRegression의 fit()`: 입력 데이터의 값들을 분석해서 모델이 데이터의 패턴을 파악하고, 예측을 위한 회귀 계수(기울기)와 절편이라는 파라미터를 '학습'합니다. 즉, 데이터의 관계를 모델에 심어주는 역할을 합니다.

특성을 변환하기 전에 필수적인 정보를 학습, 반드시 실행되어야함

`PolynomialFeatures의 fit()`: 입력 데이터의 **형태(특히 특성 개수)**만 파악하고 기억합니다. 구체적인 데이터 값에는 관심이 없고, 나중에 다항 특성을 생성할 때 몇 개의 원래 특성을 가지고 확장해야 하는지에 대한 '정보'만 '학습'합니다. **실제 특성 변환은 transform() 메서드**에서 이루어집니다.



```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures()
poly.fit([[2,3]]) # 특성 2, 3이 있다고 가정하고 특성 등 필수정보 학습
print(poly.transform([[2, 3]])) # 특성 2, 3을 실제로 변환(특성 제곱, 특성 끼리 곱)
```

# 의문점

fit() 은 특성의 개수를 파악하는 것이 전부라고 했는데, 그럼 굳이 써야하나 싶어가지고 제외 한 후 실행을 해봤다.

그런데 오류가 났다.

왜 그런지 알아 봤는데 fit()은 입력 특성 수, 생성 될 특성 수 등을 학습하는데 만약 건너뛰게 된다면 이것들을 전혀 알지 못하고 특성을 변환하는 꼴이 되는 것이다. 그래서 반드시 먼저 실행된다.

이 두 과정이 번거롭다면, 둘을 한번에 수행하는 `fit_transform()` 메소드를 활용할 수 있겠다.

#변환된 특성 중 1은 어디서 왔을까??

특성 2, 3을 활용한 변환으로 **[[1. 2. 3. 4. 6. 9.]]** 를 얻었다.

기존 특성 2, 3

특성 제곱으로 얻은 4, 9

특성끼리 곱으로 얻은 6

그런데 1은 어디서 난걸까???

다음 식을 보면 답을 얻을 수 있다.

무게 = a * 길이 + b * 높이 + c * 두께 + **d * 1**

선형 방정식의 절편 d 는 항상 1이 곱해진다고 볼 수 있다.

그래서 특성이 (길이, 높이, 두께, 1) 이라고 볼 수 있다.

사이킷런의 선형 모델은 자동으로 절편을 추가하므로 굳이 이렇게 특성을 만들 필요가 없다. 그래서 아래와 include_bias = False로 지정하여 특성을 변환할 수 있다


```python
poly = PolynomialFeatures(include_bias=False)
poly.fit([[2, 3]])
print(poly.transform([[2, 3]]))
```

```python
poly = PolynomialFeatures(include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
print(train_poly)
```
